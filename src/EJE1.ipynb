****Deteccion de outliers y pintado de los mismos sobre los valores de income****

```python
# 2. DetecciÃ³n de valores atÃ­picos en columnas numÃ©ricas
numerical_cols = ['income', 'mntwines', 'mntfruits', 'mntmeatproducts', 'mntfishproducts']

# Crear boxplots para todas las columnas numÃ©ricas
plt.figure(figsize=(20, 15))
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(6, 4, i)
    sns.boxplot(y=df_base[col])
    plt.title(f'Boxplot de {col}')
    plt.tight_layout()

plt.show()
```

```python
# DetecciÃ³n y eliminaciÃ³n de outliers para mÃºltiples columnas
numerical_cols = ['income', 'mntwines', 'mntfruits', 'mntmeatproducts', 'mntfishproducts']

# Crear una copia del dataframe para limpiar
df_clean = df_base.copy()

# FunciÃ³n para eliminar outliers usando IQR
def remove_outliers(df, column):
    Q1 = df_base[column].quantile(0.25)
    Q3 = df_base[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df_base[column] >= lower_bound) & (df_base[column] <= upper_bound)]

# Aplicar la funciÃ³n a cada columna numÃ©rica
for col in numerical_cols:
    df_clean = remove_outliers(df_clean, col)

# Crear boxplots para las columnas limpias
plt.figure(figsize=(20, 15))
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(6, 4, i)
    sns.boxplot(y=df_clean[col])
    plt.title(f'Boxplot sin outliers de {col}')
    plt.tight_layout()

plt.show

print(f"Datos originales: {df_base.shape[0]}")
print(f"Datos despuÃ©s de eliminar outliers: {df_clean.shape[0]}")
```
